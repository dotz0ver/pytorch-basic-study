## 2.6 클래스

```python
# 데이터 전처리에 관한 클래스 만들기
import numpy as np

class DataPreprocessing:
  # 클래스 내부에 외부 변수를 불러올 때는 __init__함수를 사용하고 괄호 안에 self, 변수 나열
  def __init__(self, data, target):
    # self를 붙인 변수는 클래스 내부의 함수들에서 별도의 선언 없이 자유롭게 사용 가능
    self.data = data
    self.target = target
    self.num_instances = self.data.shape[0]
    self.num_features = self.data.shape[1]
  def minmax(self):
    for i in range(self.num_features):
      col = self.data[:,i]
      # 기존 값이 스케일링 된 값으로 대체되어 self.data가 열 기준으로 계산된 0 이상 1 이하의 값을 가진 데이터로 변환됨
      self.data[:,i] = (self.data[:,i]-np.min(col))/(np.max(col)-np.min(col))
    # 단순 메서드(함수)가 호출된 인스턴스 개체에 대한 참조 반환
    return self
  def normalization(self):
    for i in range(self.num_features):
      col = self.data[:,i]
      mu, sigma = np.mean(col), np.std(col)
      self.data[:,i] = (self.data[:,i]-mu)/sigma
    return self
  # 스케일링 함수 정의
  def scaler(self, scaling=None):
    if scaling == 'minmax':
      self.minmax()
    elif scaling == 'standard':
      self.normalization()
    else:
      pass
    return self.data
```

```python
# 클래스 불러오기

data = np.random.normal(0,10,(5,5))
target = np.random.normal(0,1,5)
print(data)
data_processor = DataPreprocessing(data, target)
data = data_processor.scaler('minmax')
print(data)
```

```python
# 클래스 상속

class DataPipeline(DataPreprocessing):
  def __init__(self, data):
    self.data = data
    self.num_features = self.data.shape[1]
pipe = DataPipeline(data)
data = pipe.scaler('minmax')
```

## 2.7 그래프 그리기

```python
import numpy as np
from matplotlib import pyplot as plt

epoch = np.arange(1,100)
train_loss = 1.5/(epoch)
val_loss = .3+1/(epoch)
```

```python
# 그래프 틀 (너비, 높이)
plt.figure(figsize=(10,5))
# 리스트나 넘파이 배열 모두를 2차원 그래프로 표현
plt.plot(train_loss, 'r-*') # r-는 빨간선으로 표시, *은 해당 점은 별로 표시
plt.plot(val_loss, 'b-*')
# legend는 범례를 나타내는 함수
plt.legend(['train','validation'])
plt.title("MNIST")
plt.xlabel("epoch")
plt.ylabel("loss")
plt.show()
```
